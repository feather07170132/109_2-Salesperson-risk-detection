```python
df = pd.read_csv('AG_PROFILE_201912_FIN.csv',encoding='utf-8')
```

```python
df.head()
```

```python
df.describe()
```

```python
#轉換成str以便接下來replace方法
area_name_str = ','.join(str(x) for x in df.area_name.values) 
office_CITY_str = ','.join(str(x) for x in df.office_CITY.values) 
CLASS_str = ','.join(str(x) for x in df.CLASS.values) 
AG_GP_str = ','.join(str(x) for x in df.AG_GP.values) 
```

```python
df.area_name.unique()
```

```python
area_name_str = area_name_str.replace('01_台北一區','1').replace('02_台北二區','2').replace('03_桃竹苗地區','3').replace('04_中部地區','4').replace('05_南部地區','5')
CLASS_str = CLASS_str.replace('01_業務代表','1').replace('02_業務主任','2').replace('03_業務襄理','3').replace('04_區經理','4')
AG_GP_str= AG_GP_str.replace('01.行銷菁英','1').replace('02.銷售奇兵','2').replace('03.再購達人','3').replace('04.加油團','4')
```

```python
#回歸原本numpy陣列
area_name2 = np.array([int(x) for x in area_name_str.split(',')])
CLASS2 = np.array([int(x) for x in CLASS_str.split(',')])
AG_GP2 = np.array([int(x) for x in AG_GP_str.split(',')])
```

```python
df.area_name = area_name2
df.CLASS = CLASS2
df.AG_GP = AG_GP2
df = df.drop(columns='office_CITY')
```
