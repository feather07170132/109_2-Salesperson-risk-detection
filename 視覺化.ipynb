```python
df_grouped = df2.groupby("CLASS")
position = df_grouped["CLASS"].count()
# 繪出長條圖
plt.bar([1, 2, 3, 4], position)
plt.xticks([1, 2, 3, 4], position.index)
plt.ylabel("Number of People")
plt.xlabel("Position")
plt.title("CLASS") 
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.subplot(121) #1列2欄第一個位置
df_grouped = df3.groupby("CLASS")
position = df_grouped["CLASS"].count()

# 繪出派圖
patches, texts = plt.pie(position, 
                         labels=position.index,
                         )
plt.xlabel("全部",fontsize=15)
plt.axis("equal")
plt.title("") 
plt.subplot(122) #1列2欄第二個位置
df_grouped = df4.groupby("CLASS")
position = df_grouped["CLASS"].count()
explode = (0, 0, 0, 0)
# 繪出派圖
patches, texts = plt.pie(position, 
                         labels=position.index,
                         explode=explode)
#plt.legend(patches, position.index, loc=2, bbox_to_anchor=(1.05,1.0),borderaxespad = 0.)
plt.xlabel("有舞弊",fontsize=15)
plt.axis("equal")
plt.title("2020驗證",position=(-0.5, 1))

plt.subplots_adjust(left=0.125,
                    bottom=0.55, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.8, 
                    hspace=0.35)
plt.show()
```
